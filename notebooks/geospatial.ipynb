```python
import pandas as pd
import folium
from folium.plugins import HeatMap, MarkerCluster
import geopandas as gpd
import matplotlib.pyplot as plt

df = pd.read_parquet("../data_clean/crimes_featured.parquet")

df.head()
```

```python
df_geo = df[df["Has_Coordinates"] == True].copy()
df_geo = df_geo[["Primary_Type", "Latitude", "Longitude", "Crime_Category"]]

df_geo.shape
```

```python
chicago_center = [41.8781, -87.6298]

base_map = folium.Map(location=chicago_center, zoom_start=11)
base_map
```

```python
sample = df_geo.sample(50000, random_state=42)

heat_data = sample[["Latitude", "Longitude"]].values.tolist()

heatmap = folium.Map(location=chicago_center, zoom_start=11)
HeatMap(heat_data, radius=7, blur=6).add_to(heatmap)

heatmap
```

```python
heatmap.save("../visuals/heatmap_crime_hotspots.html")
```

```python
cluster_map = folium.Map(location=chicago_center, zoom_start=11)
cluster = MarkerCluster().add_to(cluster_map)

for _, row in sample.iterrows():
    folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=row["Primary_Type"]
    ).add_to(cluster)

cluster_map
```

```python
cluster_map.save("../visuals/crime_clusters.html")
```

```python
violent = df_geo[df_geo["Crime_Category"] == "Violent Crime"].sample(30000)

violent_map = folium.Map(location=chicago_center, zoom_start=11)
HeatMap(violent[["Latitude","Longitude"]].values, radius=7).add_to(violent_map)

violent_map
```

```python
theft = df_geo[df_geo["Primary_Type"] == "THEFT"].sample(20000)

theft_map = folium.Map(location=chicago_center, zoom_start=11)
HeatMap(theft[["Latitude","Longitude"]].values, radius=7).add_to(theft_map)

theft_map
```
